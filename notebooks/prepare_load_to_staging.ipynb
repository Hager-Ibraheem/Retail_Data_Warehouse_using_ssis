```python
customer_df = pd.read_csv("../data/raw/customers.csv")
customer_name = list(customer_df['first_name'] +" "+ customer_df['last_name']) # will be added in the sales data  
```

```python
store_df = pd.read_csv("../data/raw/stores.csv")
store_name = list(store_df['store_name']) # will be added in the sales data  
```

```python
store_df.to_csv("../data/staging/stores.csv", index=False)
```

```python
import random
random.seed(42)

def sales_update(sales_df,store_n,customer_n):
    sales_df['customer_name'] = [random.choice(customer_n) for i in range(len(sales_df))]
    sales_df['store_name'] = [random.choice(store_n) for i in range(len(sales_df))]
    return sales_df

sales_final = sales_update(sales_df,store_name,customer_name)
sales_final.to_csv("../data/staging/sales_final.csv", index=False)
```
